```python
import hashlib
from __future__ import print_function
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
hashes = [hashlib.sha256(str(i)).hexdigest() for i in xrange(10000)]
```

```python
def pow1(work, _hash):
    '''default with n leading zeros'''
    return _hash.startswith(len(str(work)) * '0')

def pow2(work, _hash):
    '''per byte less than hex value of work'''
    hex_work = hex(work)[2:]
    hex_work_len = 2
    h_prefix = _hash[:hex_work_len]

    return int(h_prefix, 16) > (int(hex_work, 16) / 4)
```

```python
work = range(1000)
```

```python
pow1_results = [len(filter(lambda h: pow1(i, h), hashes)) for i in work]
```

```python
pow2_results = [len(filter(lambda h: pow2(i, h), hashes)) for i in work]
```
